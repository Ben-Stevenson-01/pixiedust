### Loading text data from remote sources

Data files commonly reside in remote sources, such as such as public or private market places or GitHub repositories. You can load comma separated value (csv) data files using Pixiedust's `sampleData` method. 

#### Prerequisites

Import PixieDust and optionally enable the Spark Job monitor.

```python
import pixiedust
pixiedust.enableJobMonitor()
```

#### Loading  data

To load a data set invoke `pixiedust.sampleData` and specify the data set URL:

```python
homes = pixiedust.sampleData("https://openobjectstore.mybluemix.net/misc/milliondollarhomes.csv")
```

`pixiedust.sampleData` loads the data into an [Apache Spark DataFrame](https://spark.apache.org/docs/latest/sql-programming-guide.html#datasets-and-dataframes), which you can inspect and visualize using `display()`.



To inspect the automatically inferred schema and preview a small subset of the data you can use the _DataFrame Table_ view, as shown in the preconfigured example below: 


```python
display(homes)
```

To interactively explore the data set you choose a chart type and configure the options. The following bar chart has been preconfigured (`Options > Keys: CITY`, `Options > Values: PRICE` and `Options > Aggregation: AVG`) to display the average home price for each city in the data set:

```python
display(homes)
```

Changing the display **Options** you can continue to explore the loaded data set without having to pre-process the data. For example changing `Options > Key` to `YEAR_BUILT` and `Options > aggregation` to `COUNT` you can find out how old the listed properties are.

```python
display(homes)
```

To make it easy to explore the built-in visualizations, PixieDust comes with a set of curated data sets. Type `pixiedust.sampleData()` to display those data sets.

```python
pixiedust.sampleData()
```

The homes sales data set we've loaded earlier is one of the samples. Therefore we could have also loaded it by specifying the displayed data set id as parameter:

```python
home = pixiedust.sampleData(6)
```

If your data isn't stored in csv files, you can load it into a DataFrame from any supported [data source](https://spark.apache.org/docs/latest/sql-programming-guide.html#data-sources).
